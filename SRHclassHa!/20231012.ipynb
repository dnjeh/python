```python
f = open("weather2.csv", encoding="cp949")
data = csv.reader(f)
header = next(data)
data = list(data)
f.close()
```

```python
#data
header
```

```python
high = []
low = []
cnt = 0
for row in data:
    if row[2] != '':
        if "08-15" in row[0]:
            high.append(float(row[2]))
            low.append(float(row[3]))

plt.figure(figsize= (45, 10))
plt.title("   ")
plt.plot(high, "hotpink", label = "")
plt.plot(low, "skyblue", label = "")
plt.legend(loc=2)
plt.show()
```
