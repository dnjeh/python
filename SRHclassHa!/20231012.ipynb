### 광복절의 최고기온, 최저기온 그래프를 그려보자!

##### 1. 첫 번째 데이터부터 마지막 데이터까지 순서대로 읽으며 8월 15일의 데이터를 찾는다.

```python
f = open("weather2.csv", encoding="cp949")
data = csv.reader(f)
header = next(data)
data = list(data)
f.close()
```

```python
#data
header
```

```python
high = []
low = []
cnt = 0
for row in data:
    if row[2] != '':
        if "08-15" in row[0]:
            high.append(float(row[2]))
            low.append(float(row[3]))

plt.figure(figsize= (45, 10))
plt.title("광복절 최고기온 최저기온 시각화")
plt.plot(high, "hotpink", label = "최고기온")
plt.plot(low, "skyblue", label = "최저기온")
plt.legend(loc=2)
plt.show()
```
