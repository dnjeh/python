```python
f = open("weather.csv", encoding="cp949")
data = csv.reader(f)
header=next(data)
data=list(data)
f.close()
```

```python
result = []

for row in data:
    if row[2] != '':
        result.append(float(row[2]))

plt.hist(result, bins=100)
plt.show()
```

```python
aug = []
jan = []

for row in data:
    month = row[0].split('-')[1]
    if row[2] != '':
        if month == '08':
            aug.append(float(row[2]))
        if month == '01':
            jan.append(float(row[2]))

plt.hist(jan, bins=100, color='r', label="Jan")
plt.hist(aug, bins=100, color='b', label="Aug")
plt.legend()
plt.show()
```

```python
plt.figure(dpi=100)
plt.title("1월과 8월의 최고기온 히스토그램")
plt.hist(jan, bins=100, color='r', label="Jan")
plt.hist(aug, bins=100, color='b', label="Aug")
plt.legend()
plt.show()
```

```python
month = [[] for i in range(12)]
for row in data:
    if row[2] != '':
        month[int(row[0].split('-')[1])-1].append(float(row[2]))

plt.boxplot(month, showfliers=False)
plt.show()
```

```python
days = [[] for i in range(31)]
for row in data:
    if row[2] != '' and row[0].split('-')[1] == "08":
        days[int(row[0].split('-')[2])-1].append(float(row[2]))

plt.boxplot(days, showfliers=False)
plt.show()
```

```python
month = [[] for i in range(12)]
for row in data:
    if row[2] != '' and row[3] != '':
        month[int(row[0].split('-')[1])-1].append(float(row[2])-float(row[3]))

plt.boxplot(month, showfliers=False)
plt.show()
```
