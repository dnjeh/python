### 서울이 가장 더웠던 날의 최고 기온과 언제인지까지 구하기

##### 1단계: 문자열로 저장된 기온 데이터를 실수형으로 변환하기

```python
for row in data:
    row[2]=float(row[2])
```

```python
for row in data:
    for i in range(2, 4):
        if row[i]=='':
            row[i]=-999
        row[i]=float(row[i])
```

```python
for row in data[:5]:
    print(type(row[-2]))
```

##### 2단계: 최고기온이 높은 날의 날짜아 그 날의 기온값 찾기.

```python
max_temp=-999
max_date=''

for row in data:
    if max_temp < row[-2]:
        max_temp = row[-2]
        max_date = row[0]

print("기상 관측 이래 서울의 최고 기온이 가장 높았던 날은", max_date, "로,", max_temp, "도였습니다.")
```

```python
result = []
for row in data:
    if row[0].split('-')[1]=="02" and row[0].split('-')[2]=="14":
        if row[1]!='':
            result.append(float(row[1]))
print(result)
```

```python
import matplotlib.pyplot as plt

plt.title("My brithdays' avg-temp visable")
plt.plot(result, color="red", label = "Temp")
plt.legend()
plt.show()
```

```python
sum=0
for i in result:
    sum+=i
print(f"기상이 관측된 {len(result)}해 동안, 내 생일의 평균 기온은 {sum/len(result):.2f}이다.")
```

### 4. 크리스마스의 일교차의 평균은 몇 도일까?

##### 1단계, 크리스마스의 최고, 최저 기온의 차를 새 리스트에 추가한다.

```python
christmas = [] 

for row in data:
    if row[0].split('-')[1]=="12" and row[0].split('-')[2]=="25":
        if row[2]!='' and row[3]!='':
            christmas.append(float(row[2])-float(row[3]))
print(christmas)
```

```python
import matplotlib.pyplot as plt

plt.title("Christmas' avg-temp visable")
plt.plot(christmas, color="red", label = "Temp")
plt.legend()
plt.show()
```
